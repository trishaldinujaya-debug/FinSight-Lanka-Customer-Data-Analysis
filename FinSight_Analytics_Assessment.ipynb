## Data Quality
we get some real World dataset have some issues missing, duplicate, formatting problam, we want slove all those after ww can get good improve.


```python
# Store original row count
original_rows = df.shape[0]

# Remove exact duplicate rows
df = df.drop_duplicates()

# New row count
new_rows = df.shape[0]

print("Original Rows:", original_rows)
print("Rows After Removing Exact Duplicates:", new_rows)
print("Duplicates Removed:", original_rows - new_rows)
```

```python
# Find duplicate customer IDs
duplicate_customers = df[df['Customer_ID'].duplicated(keep=False)]

# Display duplicates
duplicate_customers.sort_values(by='Customer_ID')
```

```python
# Columns likely containing Yes/No values
binary_columns = [
    'Mobile_App_User',
    'Has_Loan',
    'Has_Fixed_Deposit',
    'Has_Insurance'
]

# Standardize Yes/No formatting
for col in binary_columns:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.title()

print("Binary columns standardized")
```

```python
# Standardize Province names
df['Province'] = df['Province'].astype(str).str.strip().str.title()

# Standardize District names
df['District'] = df['District'].astype(str).str.strip().str.title()

print("Province and District names standardized")
```

```python
# Invalid age records
invalid_age = df[(df['Age'] < 18) | (df['Age'] > 100)]

invalid_age[['Customer_ID', 'Age']]
```

Impossible age values were replaced useing madian valied age

```python
# Calculate median age
median_age = df[(df['Age'] >= 18) & (df['Age'] <= 100)]['Age'].median()

# Replace invalid ages with median
df.loc[(df['Age'] < 18) | (df['Age'] > 100), 'Age'] = median_age

print("Invalid ages replaced with median age:", median_age)
```

```python
# Negative savings balances
negative_savings = df[df['Savings_Balance'] < 0]

negative_savings[['Customer_ID', 'Savings_Balance']]
```

```python
# Replace negative savings with NaN
df.loc[df['Savings_Balance'] < 0, 'Savings_Balance'] = np.nan

print("Negative savings balances converted to missing values")
```

```python
# Fill missing savings balance with median
median_savings = df['Savings_Balance'].median()

df['Savings_Balance'] = df['Savings_Balance'].fillna(median_savings)

print("Missing savings balances filled with median:", median_savings)
```

```python
# Missing values after cleaning
missing_after = df.isnull().sum()

missing_after[missing_after > 0].sort_values(ascending=False)
```

```python
# Save cleaned dataset
df.to_excel("FinSight_Cleaned_Data.xlsx", index=False)

print("Cleaned dataset saved successfully")
```

## Assumption for Savings Growth Analysis
dataset does not contain monthly transaction history or month-over-month savings changes

```python
# Calculate median savings balance
median_savings = df["Savings_Balance"].median()

# Classify customers
df["Savings_Status"] = np.where(
    df["Savings_Balance"] >= median_savings,
    "Growing Savings",
    "Draining/Stagnating"
)

# Percentage calculation
savings_status = df["Savings_Status"].value_counts(normalize=True) * 100

savings_status
```

```python
# Only customers with loans
loan_customers = df[df["Has_Loan"] == "Yes"].copy()

# Create ratio
loan_customers["Debt_to_Savings_Ratio"] = (
    loan_customers["Outstanding_Loan_Balance"]
    / loan_customers["Savings_Balance"]
)

loan_customers[[
    "Customer_ID",
    "Debt_to_Savings_Ratio"
]].head()
```
